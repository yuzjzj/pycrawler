### 比較一下範例檔案中的「File I/O」與「xmltodict」讀出來的內容有什麼差異


```python
import pandas as pd
import xmltodict
with open('./data/example/sample.xml') as fd:
    r = fd.read()
print(r)

doc = dict(xmltodict.parse(r))
print(doc)
```

### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
# 1. 請問高雄市有多少地區有溫度資料？

with open("./data/example/64_72hr_CH.xml", "r") as fh:
    xml = fh.read()
xmldict = dict(xmltodict.parse(xml))
l = xmldict['cwbopendata']['dataset']['locations']['location']
len([i['locationName'] for i in l])
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度
pd.DataFrame([(
    j['locationName'],
    j['weatherElement'][0]['time'][0]['dataTime'],
    j['weatherElement'][0]['time'][0]['elementValue']['value']
) for j in l], columns=['locationName', 'dataTime', 'value'])
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度
pd.DataFrame([(
    k['dataTime'],
    k['elementValue']['value']
) for k in l[0]['weatherElement'][0]['time']], columns=['dataTime', 'value'])
```
